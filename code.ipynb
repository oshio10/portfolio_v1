````
import os
import json
import PyPDF2
from pathlib import Path

# Define file paths
pdf_path = "/data/chats/4g49g/workspace/uploads/CV Patrick Chen-2.pdf"
output_dir = Path("/data/chats/4g49g/workspace/data")
output_path = output_dir / "cv_data.json"

# Create output directory if it doesn't exist
if not output_dir.exists():
    os.makedirs(output_dir)

# Extract text from PDF
def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page in reader.pages:
            text += page.extract_text()
    return text

# Parse the extracted text into structured data
def parse_cv_text(text):
    # Create a default CV structure
    cv_data = {
        "personal": {
            "name": "Patrick Chen",  # Assuming the name from filename
            "title": "",
            "email": "",
            "phone": "",
            "location": "",
            "bio": "",
            "social": {},
        },
        "education": [],
        "experience": [],
        "skills": [],
        "projects": [],
        "languages": [],
        "interests": [],
    }
    
    # Extract sections based on common CV patterns
    lines = text.split('\n')
    current_section = ""
    
    for line in lines:
        line = line.strip()
        if not line:
            continue
            
        # Try to identify sections
        if "EDUCATION" in line.upper():
            current_section = "education"
            continue
        elif "EXPERIENCE" in line.upper() or "PROFESSIONAL" in line.upper():
            current_section = "experience"
            continue
        elif "SKILL" in line.upper():
            current_section = "skills"
            continue
        elif "PROJECT" in line.upper():
            current_section = "projects"
            continue
        elif "LANGUAGE" in line.upper():
            current_section = "languages"
            continue
        elif "INTEREST" in line.upper() or "HOBBY" in line.upper():
            current_section = "interests"
            continue
        
        # Extract contact info
        if '@' in line and current_section == "":
            cv_data["personal"]["email"] = line
        elif any(phone_indicator in line for phone_indicator in ['+', '(', ')']):
            cv_data["personal"]["phone"] = line
        
        # Add content to current section
        if current_section == "education":
            if line not in [item.get("institution", "") for item in cv_data["education"]]:
                cv_data["education"].append({"institution": line, "degree": "", "date": ""})
        elif current_section == "experience":
            if line not in [item.get("company", "") for item in cv_data["experience"]]:
                cv_data["experience"].append({"company": line, "position": "", "date": "", "description": []})
        elif current_section == "skills":
            if line not in cv_data["skills"]:
                cv_data["skills"].append(line)
        elif current_section == "projects":
            if line not in [item.get("name", "") for item in cv_data["projects"]]:
                cv_data["projects"].append({"name": line, "description": "", "technologies": []})
        elif current_section == "languages":
            if line not in [item.get("language", "") for item in cv_data["languages"]]:
                cv_data["languages"].append({"language": line, "proficiency": ""})
        elif current_section == "interests":
            if line not in cv_data["interests"]:
                cv_data["interests"].append(line)
    
    return cv_data

try:
    # Extract text from PDF
    text = extract_text_from_pdf(pdf_path)
    
    # Parse the CV data
    cv_data = parse_cv_text(text)
    
    # Save data to JSON file
    with open(output_path, 'w', encoding='utf-8') as json_file:
        json.dump(cv_data, json_file, indent=4)
    
    print(f"CV data has been extracted and saved to {output_path}")
    print("Preview of extracted data:")
    print(json.dumps(cv_data, indent=4)[:500] + "...")
    
except Exception as e:
    print(f"Error processing the CV: {str(e)}")
````
